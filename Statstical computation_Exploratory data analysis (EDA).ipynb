```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
import seaborn as sns
# Create a dictionary of car sales data
car_data = {
    'car_brand': ['Toyota', 'Honda', 'Ford', 'BMW', 'Mercedes'],
    'price': [25000, 30000, 22000, 40000, 45000],
    'mileage': [50000, 70000, 30000, 20000, 60000],
    'fuel_type': ['Petrol', 'Petrol', 'Diesel', 'Petrol', 'Diesel']
}

# Create a DataFrame from the car sales data
car_sales_df = pd.DataFrame(car_data)

# Display the dataset
print(car_sales_df)
```

```python
print("mean = ",car_sales_df.price.mean())
```

```python
print("standard_dev = ",car_sales_df.price.std())
```

```python
print("median = ",car_sales_df.price.mean())
```

```python
print(car_sales_df.car_brand.nunique())
```

```python
numeric_vars = ["price","mileage"]
numeric_stats = car_sales_df[numeric_vars].describe()
print(numeric_stats)
```

```python
categorical_vars = ['car_brand', 'fuel_type']
categorical_counts = car_sales_df[categorical_vars].nunique()
categorical_values = car_sales_df[categorical_vars].value_counts()
print(categorical_counts)
print(categorical_values)
```

```python
plt.hist(car_sales_df['price'], bins=10, edgecolor='black')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.title('Distribution of Car Prices')
plt.show()
```

```python
sns.scatterplot(x='mileage', y='price', data=car_sales_df)
plt.xlabel('Mileage')
plt.ylabel('Price')
plt.title('Relationship between Mileage and Price')
plt.show()
```

```python
average_price_by_brand = car_sales_df.groupby('car_brand')['price'].mean()
print(average_price_by_brand)

sns.barplot(x='car_brand', y='price', data=car_sales_df)
plt.xlabel('Car Brand')
plt.ylabel('Average Price')
plt.title('Average Price by Car Brand')
plt.show()
```
